```python
import cv2
import time
cap = cv2.VideoCapture('../DATA/video_capture.mp4')
fps = 25
if cap.isOpened() == False :
    print('error')
while  cap.isOpened() :
    ret,frame = cap.read()
    if ret == True :
        time.sleep(1/fps)
        cv2.imshow('frame',frame)
        if cv2.waitKey(1) & 0xFF == ord('q') :
            break
cap.release()
cv2.destroyAllWindows()
```
